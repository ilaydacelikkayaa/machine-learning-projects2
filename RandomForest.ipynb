```python
X = df.drop(['income'], axis=1)
y = df['income']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
categorical= [col for col in X_train.columns if X_train[col].dtypes == 'O']
```

```python
X_train[categorical].isnull().sum()
```

```python
X_test[categorical].isnull().sum()
```

```python
X_train['workclass'].mode()
```

```python
for i in [X_train, X_test]:
    i['workclass'] = i['workclass'].fillna(X_train['workclass'].mode()[0])
    i['occupation'] = i['occupation'].fillna(X_train['occupation'].mode()[0])
    i['native_country'] = i['native_country'].fillna(X_train['native_country'].mode()[0])
```

```python
X_train[categorical].isnull().sum()
```

```python
y_train_binary=y_train.apply(lambda x:1 if x.strip()=='>50K' else 0)
#native_country grubunda, insanların yüzde kaçının >50K maaş sınıfında olduğunu gösteriyor.
```

```python
target_means=y_train_binary.groupby(X_train['native_country']).mean()
```

```python
X_train['native_country_encoded'] = X_train['native_country'].replace(target_means)
X_train['native_country_encoded'] = X_train['native_country_encoded'].fillna(y_train_binary.mean())

X_test['native_country_encoded'] = X_test['native_country'].replace(target_means)
X_test['native_country_encoded'] = X_test['native_country_encoded'].fillna(y_train_binary.mean())
```

```python
X_train = X_train.drop("native_country", axis=1)
X_test = X_test.drop("native_country", axis=1)
```

```python
categorical
```

```python
one_hot_categories = ['workclass',
 'education',
 'marital_status',
 'occupation',
 'relationship',
 'race',
 'sex']
```

```python
from sklearn.preprocessing import OneHotEncoder #categorical to 0-1
from sklearn.compose import ColumnTransformer


encoder = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'
                              #test setinde yeni bir kategori çıkarsa hata verme
                            , sparse_output=False
                             #np.array olarak döndür
                             ), one_hot_categories)
    ],
    remainder='passthrough'  #diger column ellemesin diye
)

X_train_enc = encoder.fit_transform(X_train)
X_test_enc = encoder.transform(X_test)
```

```python
X_train_enc 
```

```python
columns = encoder.get_feature_names_out()
```

```python
columns
```

```python
X_train = pd.DataFrame(X_train_enc, columns=columns, index=X_train.index)
X_test = pd.DataFrame(X_test_enc, columns=columns, index=X_test.index)
```

```python
X_train
```

```python
cols = X_train.columns
```

```python

from sklearn.preprocessing import RobustScaler #std scaler ort=0 std=1 robustscaler->medyan=0 iqr=1
#veriyi merkeze oturtur uc degerlerden fazla etkilenmez
scaler = RobustScaler() #outliera dayanıklı

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

Eğer verinde uç değer yoksa → StandardScaler / MinMaxScaler kullanabilirsin.

Eğer verinde uç değer çoksa → RobustScaler en güvenli seçenek.

```python
X_train = pd.DataFrame(X_train, columns=[cols])
X_test = pd.DataFrame(X_test, columns=[cols])
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rnd=RandomForestClassifier(n_estimators=10,random_state=15)
```

Random Forest, aslında bir sürü Decision Tree’den (karar ağacı) oluşur.

n_estimators → Kaç tane ağaç kurulacağını belirler.

```python
rnd.fit(X_train,y_train)
```

```python
y_pred=rnd.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(f'Model accuracy score with 10 decision-trees : {accuracy_score(y_test, y_pred)}')
```

```python
rnd=RandomForestClassifier(n_estimators=100,random_state=15)
rnd.fit(X_train,y_train)
y_pred=rnd.predict(X_test)
```

```python
print(f'Model accuracy score with 10 decision-trees : {accuracy_score(y_test, y_pred)}')
```

```python
X_train = X_train.drop(["cat__workclass_ Never-worked", "cat__occupation_ Armed-Forces","cat__education_ Preschool",
                       "cat__workclass_ Without-pay", "cat__occupation_ Priv-house-serv", "cat__marital_status_ Married-AF-spouse",
                        "cat__education_ 1st-4th", "cat__education_ 5th-6th", "cat__race_ Other", "cat__education_ 12th"
                       ], axis=1)
```

```python
X_test = X_test.drop(["cat__workclass_ Never-worked", "cat__occupation_ Armed-Forces","cat__education_ Preschool",
                       "cat__workclass_ Without-pay", "cat__occupation_ Priv-house-serv", "cat__marital_status_ Married-AF-spouse",
                        "cat__education_ 1st-4th", "cat__education_ 5th-6th", "cat__race_ Other", "cat__education_ 12th"
                       ], axis=1)
```

```python
rfc = RandomForestClassifier(n_estimators=100, random_state=15)
rfc.fit(X_train, y_train)
y_pred = rfc.predict(X_test)
print(f'Model accuracy score with default decision-trees : {accuracy_score(y_test, y_pred)}')
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))

```
