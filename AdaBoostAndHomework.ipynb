```python
X = df.drop('Outcome', axis = 1)
y = df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=15)
```

```python
columns_to_fill = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
```

```python

medians = {}
for col in columns_to_fill:
    median_value = X_train[X_train[col] != 0][col].median()
    medians[col] = median_value
    X_train[col] = X_train[col].replace(0, median_value)

for col in columns_to_fill:
    X_test[col] = X_test[col].replace(0, medians[col])
```

```python
X_train.describe()
```

```python
X_test.describe()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python

ada = AdaBoostClassifier()
ada.fit(X_train, y_train)
y_pred = ada.predict(X_test)
```

```python

print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
adaboost_param = {
        "n_estimators" : [50, 70, 100, 120, 150, 200],
        "learning_rate" : [0.001, 0.01, 0.1, 1, 10]
}
```

```python
grid = GridSearchCV(estimator = AdaBoostClassifier(), param_grid = adaboost_param, cv = 5, verbose = 1, n_jobs = -1)
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
ada = AdaBoostClassifier(learning_rate=1, n_estimators=150)
```

```python

ada.fit(X_train, y_train)
y_pred = ada.predict(X_test)
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
def dene_algoritmalar(X_train, X_test, y_train, y_test):
    algoritmalar = {
        "Lojistik Regresyon": LogisticRegression(max_iter=1000),
        "Karar Aac": DecisionTreeClassifier(),
        "Rastgele Orman": RandomForestClassifier(),
        "Destek Vektr Makineleri": SVC(),
        "Naive Bayes": GaussianNB()
    }
    
    sonuclar = {}
    
    for isim, model in algoritmalar.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        
        acc = accuracy_score(y_test, y_pred)
        rapor = classification_report(y_test, y_pred, output_dict=True)
        
        sonuclar[isim] = {
            "Doruluk (Accuracy)": acc,
            "Rapor": rapor
        }
    
    return sonuclar
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
sonuclar = dene_algoritmalar(X_train, X_test, y_train, y_test)

for isim, sonuc in sonuclar.items():
    print(f"\nModel: {isim}")
    print(f"Doruluk: {sonuc['Doruluk (Accuracy)']:.2f}")
```
